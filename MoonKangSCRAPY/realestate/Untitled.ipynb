```python
import numpy as np
import pandas as pd
```

```python
realestate = pd.read_csv('mkrealestate.csv')
realestate
```

```python
realestate1 = realestate.dropna(axis=0, )
realestate1 = realestate1[realestate1.Crime_Safty!= 'NG']
realestate1 = realestate1[realestate1.Diversity!= 'NG']
realestate1 = realestate1[realestate1.Jobs_Level!= 'NG']
realestate1 = realestate1[realestate1.NightLife!= 'NG']
realestate1 = realestate1[realestate1.Outdoor_Activity!= 'NG']
realestate1 = realestate1[realestate1.PublicSchool_level!= 'NG']
realestate1 = realestate1[realestate1.Weather!= 'NG']


#df_filtered = df[(df.salary >= 30000) & (df.year == 2017)]

realestate1
```

```python
realestate1['Associate_Degree'] = realestate1['Associate_Degree'].replace({'\%':''}, regex = True).apply(pd.to_numeric, errors="coerce")
realestate1['Bachelor_Degree'] = realestate1['Bachelor_Degree'].replace({'\%':''}, regex = True).apply(pd.to_numeric, errors="coerce")
realestate1['Children_percent'] = realestate1['Children_percent'].replace({'\%':''}, regex = True).apply(pd.to_numeric, errors="coerce")
realestate1['Master_Degree'] = realestate1['Master_Degree'].replace({'\%':''}, regex = True).apply(pd.to_numeric, errors="coerce")
realestate1['Median_Rent'] = realestate1['Median_Rent'].replace({'\$':'', ',':''}, regex = True).apply(pd.to_numeric, errors="coerce")
realestate1['Median_salary'] = realestate1['Median_salary'].replace({'\$':'', ',':''}, regex = True).apply(pd.to_numeric, errors="coerce")
realestate1['Median_Home_Value'] = realestate1['Median_Home_Value'].replace({'\$':'', ',':''}, regex = True).apply(pd.to_numeric, errors="coerce")
realestate1['Numberof_reviews'] = realestate1['Numberof_reviews'].replace({'reviews':''}, regex = True).apply(pd.to_numeric, errors="coerce")
realestate1['Population'] = realestate1['Population'].replace({',':''}, regex = True).apply(pd.to_numeric, errors="coerce")



realestate1
```

```python
realestate1.dtypes
```

```python
#realvalue = realestate.sort_values('State')
realestate1['Cost_of_Living'] = realestate1['Cost_of_Living'].map({'A+': 99, 'A': 90, 'A-': 81, 'B+': 72, 'B': 63, 'B-': 54, 'C+': 45, 'C': 36, 'C-': 27, 'D+': 18, 'D': 9}).apply(pd.to_numeric, errors="coerce")

realestate1['Crime_Safty'] = realestate1['Crime_Safty'].map({'A+': 99, 'A': 90, 'A-': 81, 'B+': 72, 'B': 63, 'B-': 54, 'C+': 45, 'C': 36, 'C-': 27, 'D+': 18, 'D': 9}).apply(pd.to_numeric, errors="coerce")
realestate1['Diversity'] = realestate1['Diversity'].map({'A+': 99, 'A': 90, 'A-': 81, 'B+': 72, 'B': 63, 'B-': 54, 'C+': 45, 'C': 36, 'C-': 27, 'D+': 18, 'D': 9}).apply(pd.to_numeric, errors="coerce")
realestate1['Jobs_Level'] = realestate1['Jobs_Level'].map({'A+': 99, 'A': 90, 'A-': 81, 'B+': 72, 'B': 63, 'B-': 54, 'C+': 45, 'C': 36, 'C-': 27, 'D+': 18, 'D': 9}).apply(pd.to_numeric, errors="coerce")
realestate1['NightLife'] = realestate1['NightLife'].map({'A+': 99, 'A': 90, 'A-': 81, 'B+': 72, 'B': 63, 'B-': 54, 'C+': 45, 'C': 36, 'C-': 27, 'D+': 18, 'D': 9}).apply(pd.to_numeric, errors="coerce")
realestate1['Outdoor_Activity'] = realestate1['Outdoor_Activity'].map({'A+': 99, 'A': 90, 'A-': 81, 'B+': 72, 'B': 63, 'B-': 54, 'C+': 45, 'C': 36, 'C-': 27, 'D+': 18, 'D': 9}).apply(pd.to_numeric, errors="coerce")
realestate1['PublicSchool_level'] = realestate1['PublicSchool_level'].map({'A+': 99, 'A': 90, 'A-': 81, 'B+': 72, 'B': 63, 'B-': 54, 'C+': 45, 'C': 36, 'C-': 27, 'D+': 18, 'D': 9}).apply(pd.to_numeric, errors="coerce")
realestate1['Weather'] = realestate1['Weather'].map({'A+': 99, 'A': 90, 'A-': 81, 'B+': 72, 'B': 63, 'B-': 54, 'C+': 45, 'C': 36, 'C-': 27, 'D+': 18, 'D': 9}).apply(pd.to_numeric, errors="coerce")


realestate1
```

```python
def substring(x):
    try:
        return x[48:50]
    except:
        pass

realestate1.Review_ratings = realestate1.Review_ratings.apply(substring)#.apply(pd.to_numeric, errors="raise")



#changing to numeric

realestate1['Review_ratings'] = realestate1['Review_ratings'].apply(pd.to_numeric, errors="coerce")


```
